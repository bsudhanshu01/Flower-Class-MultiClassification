```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
columns_name=["sepal length" ,"sepal width","petal length" ,"petal width"]
```

```python
 from sklearn.tree import DecisionTreeClassifier  
```

```python
decisin_tree=DecisionTreeClassifier()
model=decisin_tree.fit(iris.data,iris.target)
```

```python
from sklearn import tree
plt.figure(figsize=(12,7))
tree.plot_tree(model,filled=True ,feature_names=columns_name)
```

```python
plt.figure(figsize=(15,7))
tree.plot_tree(model,filled=True ,proportion=True ,feature_names=columns_name)
```

```python
model.predict(iris.data)
```
